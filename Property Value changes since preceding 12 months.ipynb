```python
#Import the csv file and create the dataframe
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('C:/Users/STSC/Desktop/data.csv')
df
```

```python
df = df.drop_duplicates(subset=None, keep='first', inplace=False)
```

```python
## Removing the white spaces in the cells
df_1=df.applymap(lambda x: x.strip() if type(x)==str else x)
df_1
```

```python
## Print the list of cities
import collections
a=df_1['city']
a
seen = set()
uniq = [x for x in a if x not in seen and not seen.add(x)]  
uniq
```

```python
### Looking for the missing values of each column
df_1.info()
```

```python
## Split the Collateral column
df_1[['Current estimate 2020','Estimate for Oct 2015', 'Estimate for Oct 2019']] = df_1.Collateral.str.split(" ",expand=True,)
df_1
```

```python
## Delete Collateral column
del df_1['CoreLogic']
del df_1['Collateral']
del df_1['Quantarium']
```

```python
# Get indexes where city has value == "alameda", 'Berkley','San Francisco County'
indexNames = df_1[df_1['city'] == 'Alameda'].index
indexNames1 = df_1[df_1['city'] == 'Berkeley'].index
indexNames2 = df_1[df_1['city'] == 'San Francisco County'].index
# Delete these row indexes from dataFrame
df_1.drop(indexNames , inplace=True)
df_1.drop(indexNames1 , inplace=True)
df_1.drop(indexNames2 , inplace=True)
```

```python
## Rename the column names
df_1.columns = df_1.columns.str.replace(' ', '_')
```

```python
## Remove the $ sign from the price

def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)
```

```python
#code to remove the commas from all columns
df_1 = df_1.replace(',','', regex=True)
```

```python
#code to remove commas from just a particular column
df_1['price'] = df_1.price.replace(',','',regex=True)
df_1['total'] = df_1.total.replace(',','',regex=True)
```

```python
### Change the data tyope of 'Price' columns into float

df_1['Current_estimate_2020'] = df_1['Current_estimate_2020'].apply(clean_currency).astype('float')
df_1['Estimate_for_Oct_2019'] = df_1['Estimate_for_Oct_2019'].apply(clean_currency).astype('float')
df_1['Estimate_for_Oct_2015'] = df_1['Estimate_for_Oct_2015'].apply(clean_currency).astype('float')
df_1['total'] = df_1['total'].apply(clean_currency).astype('float')
df_1
```

```python
## Remove the rows that do not have collateral values
df_1_nan= df_1.dropna(subset=['Current_estimate_2020','Estimate_for_Oct_2019'])
df_1_nan
```

```python
from scipy import stats
#Create a box plot to remove the outliers-------------------------------------
df_1_nan.boxplot(column='total') #to see if there are outliers
```

```python
#removed the outliers with a z-score above 3.0
x=stats.zscore(df_1_nan['total'])
df_1_nan['z_score']=x
df_1_nan_cleaned = df_1_nan[df_1_nan['z_score'].abs()<=3]
df_1_nan_cleaned
```

```python
df_1_nan_cleaned
```

```python
## Calculate the price differences for the estimations and then for the actual price with the estimations
for i in df_1_nan_cleaned:
    df_1_nan_cleaned['Price_difference_2020_2019'] = df_1_nan_cleaned['Current_estimate_2020'] - df_1_nan_cleaned['Estimate_for_Oct_2019']
    df_1_nan_cleaned['Price_difference_Total_2020'] = df_1_nan_cleaned['total'] - df_1_nan_cleaned['Current_estimate_2020']
df_1_nan_cleaned
```

```python
df_1_nan.info()
```

```python
## Create the new data set
writer=pd.ExcelWriter('Dataset_new7.xlsx')
df_1_nan_cleaned.to_excel(writer)
writer.save()
```
